```python
from transformers import pipeline

classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")

cuisine_labels = [
  "Chinese",
  "Japanese",
  "Korean",
  "Vietnamese",
  "Thai",
  "Indian",
  "Nepalese",
  "Indonesian",
  "Malaysian",
  "Filipino",
  "Mongolian",
  "Pakistani",
  "Lebanese",
  "Turkish",
  "Persian",
  "Israeli",
  "Afghan",
  "Italian",
  "French",
  "Spanish",
  "Greek",
  "German",
  "Russian",
  "Portuguese",
  "Scandinavian",
  "British",
  "Irish",
  "Polish",
  "American",
  "Canadian",
  "Mexican",
  "Brazilian",
  "Argentinian",
  "Peruvian",
  "Colombian",
  "Moroccan",
  "African",
  "Jamaican",
  "Cuban",
  "Trinidadian",
  "Australian",
  "New Zealand",
  "Polynesian",
  "Hawaiian",
  "Asian-Fusion",
  "Mediterranean",
  "Vegan",
  "Vegetarian",
  "Raw",
  "Fast Food",
  "Barbecue",
  "Burgers",
  "Pizza",
  "Sandwiches",
  "Donuts",
  "Sweets/Desserts",
  "Coffeehouse",
  "Juice Bar",
  "Smoothies",
  "Bakery",
  "Brunch",
  "Healthy",
  "Street Food",
  "Fusion"
]

# Function to classify cuisine
def classify_cuisine(text):
    result = classifier(text, cuisine_labels, multi_class=False)  # Single best match
    return result['labels'][0]  # Top predicted label

# Example data
restaurants = [
    {"name": "Joe's Pizza", "address": "123 Main St", "reviews": "Best pizza in town!"},
    {"name": "Taco Loco", "address": "456 Elm St", "reviews": "Authentic Mexican tacos."},
    {"name": "Golden Dragon", "address": "789 Pine St", "reviews": "Traditional Chinese cuisine."},
]

# Combine restaurant name, address, and reviews into one input
for restaurant in restaurants:
    text = f"{restaurant['name']} {restaurant['address']} {restaurant['reviews']}"
    restaurant['cuisine'] = classify_cuisine(text)

# Print results
for restaurant in restaurants:
    print(f"Restaurant: {restaurant['name']}, Predicted Cuisine: {restaurant['cuisine']}")
```
