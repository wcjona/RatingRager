```python
from transformers import pipeline

classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")

cuisine_labels = [
  "Chinese",
  "Japanese",
  "Korean",
  "Vietnamese",
  "Thai",
  "Indian",
  "Nepalese",
  "Indonesian",
  "Malaysian",
  "Filipino",
  "Mongolian",
  "Pakistani",
  "Lebanese",
  "Turkish",
  "Persian",
  "Israeli",
  "Afghan",
  "Italian",
  "French",
  "Spanish",
  "Greek",
  "German",
  "Russian",
  "Portuguese",
  "Scandinavian",
  "British",
  "Irish",
  "Polish",
  "American",
  "Canadian",
  "Mexican",
  "Brazilian",
  "Argentinian",
  "Peruvian",
  "Colombian",
  "Moroccan",
  "African",
  "Jamaican",
  "Cuban",
  "Trinidadian",
  "Australian",
  "New Zealand",
  "Polynesian",
  "Hawaiian",
  "Asian-Fusion",
  "Mediterranean",
  "Vegan",
  "Vegetarian",
  "Raw",
  "Fast Food",
  "Barbecue",
  "Burgers",
  "Pizza",
  "Sandwiches",
  "Donuts",
  "Sweets/Desserts",
  "Coffeehouse",
  "Juice Bar",
  "Smoothies",
  "Bakery",
  "Brunch",
  "Healthy",
  "Street Food",
  "Fusion"
]
```

```python
import sqlite3
import pandas as pd

def print_table(query, p=False):
    conn = sqlite3.connect('data/ratingRager.db')
    cursor = conn.cursor()
    cursor.execute(query)

    column_names = [description[0] for description in cursor.description]

    rows = cursor.fetchall()

    column_widths = [max(len(str(row[i])) for row in rows + [column_names]) for i in range(len(column_names))]

    header = "  ".join(f"{name:<{column_widths[i]}}" for i, name in enumerate(column_names))
    if p:
        print('\n',header)
        print("-" * len(header))

        for row in rows:
            row_str = "  ".join(f"{str(item):<{column_widths[i]}}" for i, item in enumerate(row))
            print(row_str)
        print('\n')

    return pd.read_sql_query(query, conn)
```

```python
QUERY = """SELECT id, name FROM restaurants;"""
df = print_table(QUERY)
```

```python
def classify_cuisine(text):
    result = classifier(text, cuisine_labels, multi_label=False)
    print("Cuisine Type: ", result['labels'][0])
    return result['labels'][0]
```

```python
classified = {}

for index, restaurant in df.iterrows():
    print(restaurant['id'], restaurant['name'])
    if restaurant['name'] in classified:
        df.at[index, 'cuisine'] = classified[restaurant['name']]
    else:
        cuisine = classify_cuisine(restaurant['name'])
        classified[restaurant['name']] = cuisine
        df.at[index, 'cuisine'] = cuisine

    df.to_json("cuisine_classification.json", orient="records", indent=4)
```

```python
# Update score table

conn = sqlite3.connect('../data/ratingRager.db')
cursor = conn.cursor()

for index, row in df.iterrows():
    cursor.execute(
        "UPDATE scores SET cusine = ? WHERE restaurant_id = ?;",
        (row['cuisine'], row['id'])
    )

conn.commit()
conn.close()
```
