```python
import requests
import json
import os
import time
```

```python
def fetch_results(url, params):
    """
    Fetches data from the given URL with provided parameters.
    """
    time.sleep(2)
    response = requests.get(url, params=params)
    if response.status_code == 200:
        return response.json()
    else:
        print(f"Error: {response}")
        return None
```

```python
secrets_file = "secrets.json"

if not os.path.exists(secrets_file):
    raise FileNotFoundError(f"{secrets_file} not found. Please create the file with your API key.")

with open(secrets_file, "r") as file:
    secrets = json.load(file)

API_KEY = secrets.get("google_api_key")

if not API_KEY:
    raise ValueError("API key not found in secrets.json. Please add your key under 'google_api_key'.")
```

```python
place_id_file = "data/restaurant_ids_final.json"

with open(place_id_file, "r") as file:
    place_ids = json.load(file)
```

```python
place_details = {}
invalid_ids = []
```

```python
BASE_URL = "https://maps.googleapis.com/maps/api/place/details/json"

i = 0
for ids in place_ids:
    i += 1
    print("Place Number: ", i)
    PARAMS = {
        "place_id": ids,
        "fields": "name,formatted_address,rating,user_ratings_total,reviews,geometry",  # Specify only required fields to reduce payload
        "key": API_KEY,
        "reviews_sort": "newest"
    }
    stuff = fetch_results(BASE_URL,PARAMS)

    if stuff['status'] != 'OK':
        print("INVALID ID: ", ids)
        print(stuff)
        invalid_ids.append(ids)
        with open("invalid_ids.json", "w") as f:
            json.dump(invalid_ids, f, indent=4)
        continue

    place_details[i] = stuff['result']

    if 'reviews' not in stuff['result']:
        print("No Reviews for: ", ids)
        continue

    PARAMS = {
        "place_id": ids,
        "fields": "reviews",  # Specify only required fields to reduce payload
        "key": API_KEY,
        "reviews_sort": "relevant"
    }

    stuff = fetch_results(BASE_URL,PARAMS)

    if stuff['status'] != 'OK':
        print("INVALID ID: ", ids)
        print(stuff)
        invalid_ids.append(ids)
        with open("invalid_ids.json", "w") as f:
            json.dump(invalid_ids, f, indent=4)
        continue

    if 'reviews' not in stuff['result']:
        print("No Reviews for: ", ids)
        continue

    for review in stuff['result']['reviews']:
        place_details[i]['reviews'].append(review)

    with open("restaurant_reviews.json", "w") as f:
        json.dump(place_details, f, indent=4)
```
